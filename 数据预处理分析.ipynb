```python
import pandas as pd
import numpy as np
#数据读取
train = pd.read_csv('data/train.csv',encoding='gbk')
test = pd.read_csv('data/test.csv',encoding='gbk')
```

```python
#合并训练集和测试集，方便进行数据处理
test['患有糖尿病标识']=-1
data = pd.concat([train,test],axis=0,ignore_index=True)
#统计数据缺失情况(百分比)
null = data.isnull().sum()/len(data)*100
null
```

```python
#考虑数据为零的情况，看有多少数据为零(主要看口服耐糖量测试、胰岛素释放实验、肱三头肌皮褶厚度)
data = data.replace(0,np.nan)
zero = data.isnull().sum()/len(data)*100
zero
```

```python
#把替换的数据在重新转换回来
data = data.replace(np.nan,0)
data.head(5)
```

```python
#对糖尿病家族史进行转换
dict1 = {'无记录':0,'叔叔或姑姑有一方患有糖尿病':1,'叔叔或者姑姑有一方患有糖尿病':1,'父母有一方患有糖尿病':2,}
data['糖尿病家族史'] = data['糖尿病家族史'].map(dict1).astype('category')
#对体重指数进行转换：低于18.5过轻，24-27超重，27-32肥胖，32+非常肥胖
def BMI(a):
    if a<18.5:
        return 0
    elif 18.5<=a<=24:
        return 1
    elif 24<=a<=27:
        return 2
    elif 27<=a<=32:
        return 3
    else:
        return 4
data['BMI'] = data['体重指数'].apply(BMI)
data['出生年份'] = 2022 - data['出生年份']
data['性别'] = data['性别'].astype('category')
data.head(5),data.dtypes
```

```python
#可以从性别、年龄、体重指数、糖尿病家族史、患有糖尿病标识等多个角度，对数据进行分组聚合，然后在每一组中观察舒张压、口服耐糖量测试与患有糖尿病的规律/关系
for tmp_df in data.groupby('出生年份'):
    #分组聚合：key,表格
    print(tmp_df[0],tmp_df[1].shape)
```

```python
#对分组后的表格，每一组的舒张压聚合----最值
data.groupby("性别")["舒张压"].max()
```

```python
#每一组的取值情况
data.groupby("出生年份")["舒张压"].value_counts()
```

```python
#同时聚合多个指标
data.groupby("患有糖尿病标识")["舒张压"].agg(['mean','min','max'])
```

```python
#同时聚合多个指标
data.groupby("患有糖尿病标识").agg({
    '舒张压':['mean','min','max'],
    '口服耐糖量测试':['mean','min','max'],
})
```
